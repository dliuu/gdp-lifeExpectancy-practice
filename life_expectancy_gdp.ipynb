```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('all_data.csv')

print(df.head(10))
df.describe(include='all')
```

```python
print(df.GDP.mean())
countries = df.Country.unique()
print(countries)

#all_data
sns.lmplot(x='Year', y='GDP', data=df)
plt.show()
plt.clf()

sns.lmplot(x='Year', y='Life expectancy at birth (years)', data=df)
plt.show()
plt.clf()
```

```python
for country in df.Country.unique():
    print(country)
    sns.lmplot(x='Year', y='GDP', data=df[df.Country == country])

    plt.title('GDP Over Time, ' + country)
    plt.xlabel('Year')
    plt.xticks(df[df.Country == country]['Year'])
    plt.ylabel('GDP')
    plt.yticks(df[df.Country == country]['GDP'])

    plt.show()
    plt.clf()
```

```python
for country in df.Country.unique():
    print(country)
    sns.lmplot(x='Year', y='Life expectancy at birth (years)', data=df[df.Country == country])

    plt.title('Life Expectancy Over Time, ' + country)
    plt.xlabel('Year')
    plt.xticks(df[df.Country == country]['Year'])
    plt.ylabel('Life expectancy at birth (years)')
    plt.yticks(df[df.Country == country]['Life expectancy at birth (years)'])

    plt.show()
    plt.clf()
```

```python
for country in df.Country.unique():
    print(country)
    sns.lmplot(x='GDP', y='Life expectancy at birth (years)', data=df[df.Country == country])

    plt.title('GDP and Life Expectancy, ' + country)
    plt.xlabel('GDP')
    plt.xticks(df[df.Country == country]['GDP'])
    plt.ylabel('Life expectancy at birth (years)')
    plt.yticks(df[df.Country == country]['Life expectancy at birth (years)'])

    plt.show()
    plt.clf()
```

```python
for country in df.Country.unique():
    meanExpectancy = df[df.Country == country]['Life expectancy at birth (years)'].mean()
    print(country + ': ' + str(meanExpectancy))
```

```python
for country in df.Country.unique():
    print(country + ': ')
    sns.histplot(data=df[df.Country == country], x=df['Life expectancy at birth (years)'])
    
    plt.title('Distribution of Life Expectancy, ' + country)
    
    plt.show()
    plt.clf()
```
